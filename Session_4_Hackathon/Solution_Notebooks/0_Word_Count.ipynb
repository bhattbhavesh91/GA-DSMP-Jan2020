```
import pandas as pd
def word_count(inp_str):
    counts = dict()
    inp_str = inp_str.lower()
    words = inp_str.split()
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    print(counts)
    df = pd.DataFrame(counts.items(), columns=['Word', 'Word_Count'])
    return df
```

```
word_count("My name is Bhavesh bhavesh")
```
